```python
from keras.models import Sequential
from keras.layers import Input, Dense
from keras.utils import to_categorical
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
model = Sequential()
model.add(Dense(10, input_dim=14, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['acc'])

model.summary()
```

```python
data = pd.read_csv('train.csv',index_col=0)
```

```python
x_train = data[['MACD','SIGNAL LINE','+DI','-DI',
                'adx','storsi','%K','%D',
                'aroon_up','aroon_down','ATR','CCI',
                'OBV','AO']]
x_train = x_train.to_numpy()
data['Signal'] = np.where(data['Signal']=='buy',0,(np.where(data['Signal']=='sell',1,2)))
y_train = data['Signal'].to_numpy()
y_train = to_categorical(y_train)
```

```python
model.fit(x_train, y_train, epochs=500, batch_size=5)
```

```python
model.save('my_model.h5')
```

```python
testdata = pd.read_csv('test.csv')
x_test = testdata[['MACD','SIGNAL LINE','+DI','-DI',
                'adx','storsi','%K','%D',
                'aroon_up','aroon_down','ATR','CCI',
                'OBV','AO']]
x_test = x_test.to_numpy()
testdata['Signal'] = np.where(testdata['Signal']=='buy',0,(np.where(testdata['Signal']=='sell',1,2)))
y_test = testdata['Signal'].to_numpy()
```

```python
y_pred = model.predict(x_test)
y_pred = np.argmax(y_pred,axis = -1)
# print(y_test)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix")
print(cm)
```
